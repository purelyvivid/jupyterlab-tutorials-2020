```python
from PIL import Image, ImageOps
from PIL import ImageEnhance
from PIL import ImageFilter
import numpy as np
```

# 打開圖檔

從網站

```python
import requests
from io import BytesIO
```

```python
img_url = "https://i.imgur.com/Keno2OW.png"
response = requests.get(img_url)
img = Image.open(BytesIO(response.content)).convert("RGB")
```

# 濾鏡

```python
img_ = img.filter(ImageFilter.BLUR)#模糊化
img_
```

```python
img_ = ImageEnhance.Contrast(img).enhance(3.0)#對比增強
img_
```

```python
img_ = img.filter(ImageFilter.FIND_EDGES) #邊緣增強
img_
```

```python
img_ = ImageEnhance.Sharpness(img).enhance(3.0) #銳度增強
img_
```

```python
img_ = img.filter(ImageFilter.SMOOTH)  #平滑
img_
```
